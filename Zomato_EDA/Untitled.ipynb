```python
#now we merge the new dataset with main dataset
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#merging 2 dataset
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts() #we can say the zomato most available in india
```

```python
#here we save the name of country and values in variables.
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart of top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Zomato maximum records or transaction are from India after that USA and then United Kingdoms.

```python
final_df.columns
```

```python
#here we perform groupby with some column
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

Obeservation:
1. When rating between 4.5 and 4.9 --->  Excellent
2. When rating between 4.0 and 4.4 --->  Very good
3. When rating between 3.5 and 3.9 --->  Good
4. When rating between 2.5 and 3.4 --->  Average
5. When rating between 1.8 and 2.4 --->  Poor
6. When rating 0.0                 --->  Not rated

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

```python
# now plotting a barplot 
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Obeservation:
1. 'Not Rated' count is very high.
2.  Maximum number of ratings are between 2.6 to 4.2


```python
# plotting count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

in this countplot on the y axis frequency of color count is given
'Orange' color has more count 
white color which is present by blue has low count
